```python
from lxml import etree;
from PIL import Image
import xmltodict, cv2, os
import xml.etree.ElementTree as ET
```

```python
#xml 파일 불러옴
xml_file = 'C:/Users/chltp/Downloads/이상행동 CCTV 영상/01.폭행(assult)/insidedoor_01/10-1/10-1_cam01_assault03_place07_night_spring.xml'
doc = ET.parse(xml_file)
root = doc.getroot()
dic = []
action_list = []
action_name = []

action_list=root.findall("./object/action")
for i in action_list:
    for frame in i.findall("frame"):
        dic.append([i.find("actionname").text, int(frame.find("start").text), int(frame.find("end").text)])
for action in root.iter("action"):
    action_name.append(action.find("actionname").text)
    
print(dic)
print(action_name)
```

```python
vidcap = cv2.VideoCapture('C:/Users/chltp/Downloads/이상행동 CCTV 영상/01.폭행(assult)/insidedoor_01/10-1/10-1_cam01_assault03_place07_night_spring.mp4')
success,image = vidcap.read()

count = 1
success = True

for i in action_name:
    try: #name 폴더 만들기
        print("./" + i)
        os.makedirs("./" + i)
    except:
        pass

while success:
    success,image = vidcap.read()
    if(int(vidcap.get(1)) % 1 == 0):
        #print('Saved frame number : ' + str(int(vidcap.get(1))))
        count += 1
        for i in dic:
            if i[1]<= count and count <= i[2]:
                cv2.imwrite("./"+i[0]+"/%d.jpg" % count, image)
                print('Saved frame%d.jpg' % count)
            else: continue
        
vidcap.release()
```
